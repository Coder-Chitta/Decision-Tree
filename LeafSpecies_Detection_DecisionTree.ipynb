## Importing Necessary Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Dataset

```python
from sklearn.datasets import load_iris
dataset = load_iris()
```

## Summerize Dataset

```python
dataset.data 
```

```python
dataset.target
```

```python
dataset.data.shape
```

## Segregating Data into X (Independent Variable) & Y (Dependent Variable)

```python
X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
X
```

```python
Y = dataset.target
Y
```

## Splitting Dataset into Train & Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.25, random_state=0)
print(X_train.shape)
print(X_test.shape)
```

## Finding best max_depth Value

```python
accuracy = []
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

for i in range(1,10):
    model = DecisionTreeClassifier(max_depth=i, random_state=0)
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    score = accuracy_score(y_test, pred)
    accuracy.append(score)
    
plt.figure(figsize=(14,6))
plt.plot(range(1,10), accuracy, color='red', linestyle='dashed', marker='o', markerfacecolor='blue', markersize='10')
plt.title("Finding Best max_depth")
plt.xlabel("pred")
plt.ylabel("score")
plt.show()
```

## Training

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
model.fit(X_train,y_train)
```

## Visualization

```python
from sklearn.tree import plot_tree

# Train a decision tree classifier on the Iris dataset
model = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
model.fit(X_train, y_train)

# Visualize the decision tree
plt.figure(figsize=(20,10))
plot_tree(model, feature_names=dataset.feature_names, class_names=dataset.target_names, filled=True)
plt.show()
```

## Prediction

```python
y_pred = model.predict(X_test)
y_pred
```

```python
## To see Actual and Predicted Values
print(np.concatenate((y_test.reshape(len(y_test),1), y_pred.reshape(len(y_pred),1)),1))
```

## Accuracy

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
confusion_matrix(y_test, y_pred)
```

```python
#Calculate classification report
class_report = classification_report(y_test, y_pred)
print('Classification Report:\n', class_report)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
